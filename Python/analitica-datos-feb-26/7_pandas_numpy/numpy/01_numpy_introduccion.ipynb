# 1. Introducción a Numpy

Numpy (Numerical Python) es una libreria de python que se usa en ciencia de datos. Introduce el concepto de nparray es una estructura de datos mucho más eficiente que las listas nativas de python, sobre todo para trabajar con datos numéricos

## ¿Por qué NumPy en lugar de listas nativas?

Las lista nativas son flexible pero muy lentas se trabaja elemento a elemento. No estan pensandas y optimizadas para calculos matematicos. Numpy almacena datos siempre del mismo tipo (siempre numericos). Permite operaciones vectorizadas mucho mas rapidas

```python
import time

# sumar de 1 millón de elementos de una lista python
lista = list(range(1_000_000))
inicio = time.time()
resultado = [ x * 2 for x in lista ]
print('tiempo',  time.time() - inicio)
```

```python
# la misma operacion pero con numpy
array = np.arange(1_000_000)
inicio = time.time()
resultado = array * 2
print('tiempo', time.time() - inicio)
```

# El objeto ndarray

Un `ndarray` es un array N-dimensional de elementos del **mismo tipo**. A diferencia de una lista, todos sus elementos ocupan el mismo espacio en memoria, lo que permite operaciones vectorizadas sin bucles explícitos.

```python
np_lista = np.array([10,20,30,40,50])
lista = [1,2.5,3,4,5]
print(lista)

print(np_lista)
print(type(np_lista))
# puede saber el tipo de todos los datos mi array
print( np_lista.dtype )

np_dobles = np_lista * 2
print(np_dobles)
```

```python
booleanos = np.array([1,0,1,0,1,0], dtype=bool)
print(booleanos)
print(booleanos.dtype)
```

# Creacion de array con NumPy

NumPy ofrece multiples forma de crear arrays, desde datos ya existentes (lista, tuplas, fichero) o mediante funciones generadoras.

```python
# Vector de zeros
zero_vector = np.zeros(4) # un registro con 3 ceros
print(zero_vector)

# matriz de ceros 4x3 => 2x3 ....
matriz_ceros = np.zeros((4, 3)) # 4 filas y 3 columnas
print(matriz_ceros)

# matriz de unos
matriz_unos = np.ones((2,3))
print(matriz_unos)

# matriz de cualquier valor
matriz_7 = np.full((3,3), 7)
print(matriz_7)

print('-' * 30)
# matriz identidad.
identidad = np.eye(4)
print(identidad)

```

# Creacion de arrays aleatorios

Imprescindibles para hacer simulaciones, inicializacion de pesos ML, pruebas etc

semilla o seed, fija la aleatoriedad de los numeros dentro de mi array o vector, cuanto mas grande sea el numero mayor aleatoriedad (seed 0-100)

```python
# fijamos la semilla
np.random.seed(50) # fija la aleatoriedad para obtener siempre los mismos valores.
alturas = np.round(np.random.uniform(1.40, 2.16, size=10), 2)
print(alturas)
```
