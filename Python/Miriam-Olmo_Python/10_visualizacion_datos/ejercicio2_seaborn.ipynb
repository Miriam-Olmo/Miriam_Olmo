```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Auditoria Grafica de aplicación de gimnasio Code and Fit

Sois los analistas de datos de una nueva app de entrenamiento y programación. Vuestra misión es generar un reporte visual utilizando seaborn para que la directiva pueda entender el estado de la empresa. Tenéis 4 misiones que corresponden a 4 datasets distintos.

```python
# Dataset 1: Crecimiento de usuarios por mes y tipo de plan
datos_crecimiento = {
    "Mes": ["Ene", "Feb", "Mar", "Abr", "May", "Ene", "Feb", "Mar", "Abr", "May"],
    "Usuarios_Activos": [500, 800, 1200, 1500, 2100, 200, 350, 600, 900, 1400],
    "Plan": ["Gratis", "Gratis", "Gratis", "Gratis", "Gratis", "Premium", "Premium", "Premium", "Premium", "Premium"]
}

# Dataset 2: Puntuación de soporte técnico por área y plataforma
datos_soporte = {
    "Area": ["App Core", "App Core", "Pagos", "Pagos", "Contenido", "Contenido"],
    "Satisfaccion": [8.1, 7.5, 9.0, 8.8, 9.5, 9.2],
    "Plataforma": ["iOS", "Android", "iOS", "Android", "iOS", "Android"]
}


# Dataset 3: Comportamiento de usuarios (Tiempo de uso vs Ejercicios completados)
datos_comportamiento = {
    "Horas_Uso_Mensual": [5, 12, 8, 20, 15, 3, 10, 25, 18, 6],
    "Retos_Completados": [2, 8, 5, 15, 10, 1, 6, 20, 14, 3],
    "Nivel_Usuario": ["Junior", "Senior", "Junior", "Senior", "Senior", "Junior", "Junior", "Senior", "Senior", "Junior"]
}

# Dataset 4: Tiempo de carga de la aplicación en milisegundos
tiempos_carga = [120, 125, 130, 135, 140, 142, 145, 148, 150, 155, 158, 160, 165, 170, 180, 132, 138, 144, 152, 162]

sns.set_theme(style='darkgrid')  # configurar el tema 
```

```python
### Misión 1 (Líneas): Evolución del Crecimiento.
# Usa sns.lineplot con el crecimiento.

# Separa por "Plan" usando color (hue) y estilo de trazo (style).

# Añade marcadores a los puntos (markers=True) y haz que la línea sea gruesa (linewidth=2.5).

df_crecimiento = pd.DataFrame(datos_crecimiento)
plt.figure(figsize=(10,6))
sns.lineplot(data=df_crecimiento, 
x='Mes', 
y='Usuarios_Activos', 
hue='Plan', 
style='Plan', 
markers=['o', 'D'], 
markersize=10, 
linewidth=2.5)
plt.title('evolucion mensual de crecimiento de usuarios')
plt.show()
```

```python
### Misión 2 (Barras): Análisis de Soporte.
# Usa sns.barplot con el soporte para ver la "Satisfaccion" por "Area".
# Separa por "Plataforma" (hue).
# Aplica una paleta de colores personalizada de tu elección, pon un borde negro a las barras (edgecolor="black") y un poco de transparencia (alpha=0.8).

df_soporte = pd.DataFrame(datos_soporte)

plt.figure(figsize=(10,6))
plt.title('satisfaccion en soporte por area y plataforma')
plt.ylim(0,10)
sns.barplot(data=df_soporte, 
x='Area', 
y='Satisfaccion', 
hue='Plataforma', 
palette=['blue', 'violet'], 
edgecolor='black', 
linewidth=1.8,
alpha=0.8)
plt.show()
```

```python
### Misión 3 (Dispersión): Comportamiento de Usuarios.
# Usa sns.scatterplot con el comportamiento para comparar "Horas_Uso_Mensual" vs "Retos_Completados".

# Diferencia el "Nivel_Usuario" tanto por color (hue) como por forma (style).

# Haz que el tamaño de todos los puntos sea grande (ej: s=120) y ponles un borde oscuro (edgecolor="black").

df_comportamiento = pd.DataFrame(datos_comportamiento)
plt.figure(figsize=(10,6))
plt.title('relacion horas de uso vs retos completados')
sns.scatterplot(data=df_comportamiento, x='Horas_Uso_Mensual',
y='Retos_Completados',
hue='Nivel_Usuario',
style='Nivel_Usuario',
s=120,
edgecolor='black',
linewidth=0.5,
palette=['darkblue', 'blue'],
markers={'Junior': 'o', 'Senior': 'D'})
plt.grid()
plt.show()
```

```python
### Misión 4 (Histograma): Rendimiento de la App.

# Usa sns.histplot con el tiempos
# Muestra la curva de densidad (kde=True).

# Cambia el color a un tono morado o azul oscuro, usa unos 8 o 10 bins (analiza si lo que os pido es correcto), y dale un toque distinto cambiando el estilo visual (element="step" o modificando el grosor del borde).

df_tiempos = pd.DataFrame({'ms_carga': tiempos_carga})
plt.figure(figsize=(10,6))
plt.title('distribucion de tiempos de carga de la app')
hist = sns.histplot(data=df_tiempos, x='ms_carga', kde=True, bins='auto', color='indigo', linewidth=2, alpha=0.2, line_kws={'linewidth': 3, 'linestyle': '--'})
hist.lines[0].set_color('black')
plt.grid()
plt.show()
```
