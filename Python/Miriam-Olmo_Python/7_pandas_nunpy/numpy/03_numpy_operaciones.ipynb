# operaciones con numpy

en este tema vamos a aprender como operar matematicamente sobre arrays sin tener que recorrerlos. sin bucles. esto es uno de los conceptos mas importantes de numpy.

```python
examen_matematicas = [3.5, 5, 6.5, 2, 10, 9]
examen_lengua = [6.5, 1, 3, 4, 7, 8]

ex_mat_np = np.array(examen_matematicas)
ex_lng_np = np.array(examen_lengua)

print( (ex_mat_np + ex_lng_np) / 2)
```

```python
a = np.array([10,20,30,40,50,60])
b = np.array([2,4,6,8,10,12])

print( 'a + b', a + b )
print( 'a - b', a - b )
print( 'a * b', a * b )
print( 'a / b', a / b )
print( 'a ** 2', a ** 2 )
print( 'b % 3', b % 3 )
print( 'dobles', b * 2 )
```

## broadcasting: cuando operamos con elementos de distinta longitud

```python
ap = np.array([ [20,15,60],
                                    [30,45,39.90],
                                    [10,12.90,30],
                                    [20,15.30,60]])

le = np.array([12,3,8.9])

resultado = ap + le
print(resultado)
```

## Operaciones de comparacion

Las comparaciones (`>`, `<`, `==`, `!=`, `>=`, `<=`) también se aplican elemento a elemento y devuelven un array de `True`/`False`:

```python
notas = np.array([4,7,6,3,5,10,8,7,9])

# ¿quien aprobaria?
aprobados = notas >= 5
print(notas)
print(aprobados)

# preguntas globales
print('¿Algún alumno suspende?', np.any(notas < 5)) # True
print('¿Todos aprueban?', np.all(notas >= 5)) # False
print('¿Cuanto alumnos aprueban?', np.sum(aprobados)) # True=1 False=0 sum el conteo de 1
print('¿Cuanto alumnos suspende?', np.sum(notas < 5)) 
```

## funciones especiales => ufincs

estamos hablando de operaciones como: max, min, raiz cuadrada, exponenciales, logaritmicas, sin, cos, redondeos, valores absolutos.

```python
numeros = np.array([-2,1.3,0.8,-5.3, 1.0, 7, 29])

# sacar valor absoluto
numeros_abs = np.abs(numeros)
print(numeros_abs)

# exponencial - b^e: Crecimiento de poblaciones, interes compuesto, magnitudes.
# si invierte 1000€ al 5% anual continuo, tres n años tienes: 1000 * e^(0.05*n)

anios = np.array([0,1,5,10,20,40])
capital = np.round(10000 * np.exp(0.02 * anios), 2) 
print(capital)


# logaritmo es la inversa de la exponecial, escala de Richter
# Escala de Ritcher cada grado = 10x energía
# magnitud = log10(energia)

energias = np.array([10, 100, 1000, 10000, 100000, 1000000])
magnitudes = np.log10(energias)
print(magnitudes)
# representacioj grafica de magnitudes y energias
for e, m  in zip(energias, magnitudes):
    print(f"Energia: {e} -> Magnitud: {m}")

# raiz cuadrada
resultado = np.sqrt([4, 9, 16, 25, 36, 49, 64, 81])
print(resultado)

# redondeos
numeros_racionales = np.array([2.456, 3.6543, 67.234,9.2345,10.345635])

print(np.round(numeros_racionales, 2))

# redondeo a la baja
print(np.floor(numeros_racionales))

# redondeo a la alta
print(np.ceil(numeros_racionales))

# truncar (trunc): trunca el numero eliminando parte del decimal, siempre hacia cero
print(np.trunc(numeros_racionales))
```
