## La media (promedio)
La media es el valor central o tipico de un conjunto de datos. Se calcula sumando todos los valores del conjunto y dividiendolo entre el numero total de elemento.

$$\bar{x} = \frac{x_1 + x_2 + \cdots + x_n}{n} = \frac{\sum_{i=1}^{n} x_i}{n}$$

**Ejemplo:** Las notas de un alumno son [5, 7, 8, 6, 9]. La media es:
$$\bar{x} = \frac{5 + 7 + 8 + 6 + 9}{5} = \frac{35}{5} = 7$$

````python
notas = np.array([5, 6, 2, 7, 1, 8, 9])

# python
media_python = sum(notas) / len(notas)
print(media_python)

# numpy
print('media numpy: ', np.mean(notas))
````

# la mediana

la mediana es el valor que queda justo en el centro cuando ordenamos los datos de menor a mayor. la mitad de los datos estan por encima y la otra mitad por debajo.

    - si los datos son un numero impar: la mediana es justo el valor central.
    - Si los datos son pares: La media es la media de los dos valores centrales.

**Ejemplo con 5 valores** (impar):
```
Datos: [3, 1, 7, 5, 2]  →  Ordenados: [1, 2, 3, 5, 7]
                                              ↑
                                        Mediana = 3
```

**Ejemplo con 6 valores** (par):
```
Datos: [4, 1, 7, 5, 2, 8]  →  Ordenados: [1, 2, 4, 5, 7, 8]
                                                  ↑↑
                                          Mediana = (4+5)/2 = 4.5
```

La mediana **no se deja engañar por valores extremos** (al contrario que la media).

````python
# usando salarios vamos a calcular la mediana
print( 'Media', np.mean(salarios) )
print( 'Mediana', np.median(salarios) ) 
# ordena con numpy
print( 'salarios ordenados', np.sort(salarios, descending=False) )
````

## La varianza y la desviación tipica.

La media y mediana nos dicen el valor central, pero no nos dicen **cuanto varian ** los datos alrededor de ese valor. Para eso usamos la varianza y la desviación típica

** Ejemplo ** Dos clases de alumnos tiene la misma nota media 6, pero:
Clase A: [6,6,6,6,6] -> todos igual, sin variación
Clase B: [1,4,6,8,9] -> muy dispersa, gran variacion

## La varianza
Mide la dispersión media de los datos respecto de la media. Se calcula asi.

 - Calcular la media:  $\bar{x}$
 - Resta la media a cada dato: $(x_i - \bar{x})$
 - Elevamos al cuadrado para que no se cancelen los positivos y negativos: $(x_i - \bar{x})^2$
 - Calcular le media de los cuadrados

$$\text{Varianza} = \frac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}$$

````python
clase_A = np.array([6, 6, 6, 6, 6])
clase_B = np.array([1, 4, 6, 8, 9])

print(np.var(clase_A))
print(np.var(clase_B))
````

# desviacion tipica

es simplemente la raiz cuadrada de la varianza. se usa mas que la varianza, porque tiene las mismas unidades que los datos originales.

$$\sigma = \sqrt{\text{Varianza}}$$

````python
print('desviacion tipica ', np.std(clase_A)) # dispersion 0, no hay varianza
print('desviacion tipica ', np.std(clase_B)) # 2.8 con respecto a la media
````

````python
print('clase_A')
print('media ', np.mean(clase_A))
print('mediana ', np.median(clase_A))
print('varianza ', np.var(clase_A))
print('desviacion tipica', np.std(clase_A))

print('-'*30)

print('clase_B')
print('media ', np.mean(clase_B))
print('mediana ', np.median(clase_B))
print('varianza ', np.var(clase_B))
print('desviacion tipica', np.std(clase_B))
````

````python
# desviacion típica para este array
datos = np.array([4,7,8,5,6])
media = np.mean(datos)

diferencias = datos - media
print(diferencias)

cuadrados = diferencias ** 2
print(cuadrados)

varianza = np.mean(cuadrados)
print(varianza)

desv_tipica = np.sqrt(varianza)
print(desv_tipica)
````

## maximo, minimo y rango

son conceptos simples pero muy usados.
    - **minimo**: el valor mas pequeño del conjunto
    - **maximo**: el valor mas grande del conjunto
    - **rango**: la diferencia entre el maximo y el minimo. mide cuanto se extienden los datos
  $$\text{Rango} = \text{Máximo} - \text{Mínimo}$$

````python
temperaturas = np.array([15.2,18.5,22.1,25.8,28.3,26.0, 23.4, 19.7,16.1, 14.8])

print('minimo ', np.min(temperaturas))
print('maximo ', np.max(temperaturas))
print('rango ', np.ptp(temperaturas), 'cº (max-min)')
print('media ', np.mean(temperaturas))

print('dia de la lista mas frio ', np.argmin(temperaturas))
print(' dia de la lista mas caluroso ', np.argmax(temperaturas))
````

## percentiles y cuartiles

un percentil (p)  es el valor por debajo de p% de datos.

**ejemplo**: si tu nota en un examen esta en el percentil 80, significa que el 80% de los estudiantes han sacado una nota igual o inferior a la tuya. estas en el 20% mejor de la clase.

los cuartiles dividen la muestra en cuatro partes.
    -  minimo Q1
    - Q2
    - Q3
    - maximo Q4

    ```
Mínimo       Q1.      Q2.     Q3       Máximo
  | --------- | ------ | ----- |-------- |
        25%.      25%.     25%.    25%
```
    - Q1 (precentil 25): el 25% de los datos estanpor debajo
    - Q2 (percentil 50): la mediana, el 50% por debajo
    - Q3 (percentil 75): el 75% de la muestra esta por debajo

El **rango intercualitativo** (IQR) =  Q3 - Q1

Mide la dispersión del 50% central de los datos. Es más robusto para el cálculo de la desviación típica por que ignora los extremos.

````python
notas_clase = np.array([3,4,4,5,5,5,6,6,6,6,7,7,7,7,7,8,8,8,8,8,8,9,9,10])

Q1 = np.percentile(notas_clase, 25)
Q2 = np.percentile(notas_clase, 50) # mediana
mediana = np.median(notas_clase)
Q3 = np.percentile(notas_clase, 75)
IQR = Q3 - Q1

print('Q1 25% de los alumnos estan por debajo de esta nota', Q1)
print('Q2 50% de los alumnos estan por debajo de esta nota', Q2)
print('Mediana', mediana, 'tiene que ser igual que el anterior')
print( 'Q3 75% de los alumnos estan por debajo de esta nota', Q3 )
print('IQR', IQR)

# en que percentil esta un alumno con una nota de 8
nota_alumno = 8
calculo = (np.sum(notas_clase <= nota_alumno)/ len(notas_clase))
percentil = calculo * 100
cuartil = percentil / 25
print('percentil', np.ceil(percentil))
print('cuartil',  np.ceil(cuartil ))
print('decil', np.ceil(percentil/10))
````

## sumas acumuladas ¡¡¡¡IMPORTANTE!!!!

la suma acumulada va sumando valores uno a uno, guardando el total de cada posicion. es util para ver como crece un total a lo largo del tiempo(ventas acumuladas, km recorridos, ect...)

ventas_diarias = np.array([100, 150, 80, 200, 120, 90, 175])
acumulado = [100, 250, 330, 530, 650, 740, 915]

````python
ventas_diarias = np.array([100, 150, 80, 200, 120, 90, 175])
dias = np.array(['Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom'])
total = np.sum(ventas_diarias)
acumulado = np.cumsum(ventas_diarias)
print('Dia    | Ventas       | Total acumulado   ')
print('-'*40)
for dia, v, ac in zip(dias, ventas_diarias, acumulado):
    print(f'{dia:6s}.    | {v:6d}.    | {ac:4d}.    ')
````

## parametro axis: manejar array de n-dimension por ejes

axis, cuando tengo minimo de dos dimensiones(matriz), puedess querer:
    - la media de cada columna
    - la media de cada fila

el parametro **axis** indica en que direccion colapsan los datos.

    - axis = 1 recorre las columnas -> da 1 por fila(opera hacia la derecha)
    - axis = 0 recorre las filas -> da 1 por columna(opera hacia abajo)

   

````python
notas = np.array([
        [6,7,8], # juan
        [4,5,7], # ana
        [4,5,9], # paula
        [2,7,4] # manolo
        ])

media_asignatura = np.mean(notas, axis=0)

print('mates ', media_asignatura[0])
print('lengua ', media_asignatura[1])
print('ingles ', media_asignatura[2])
````

**Resumen del tema:**

| Concepto | Fórmula | `numpy` | Para qué sirve |
|----------|---------|---------|----------------|
| Media | $\bar{x} = \frac{\sum x_i}{n}$ | `np.mean()` | Valor central típico |
| Mediana | valor del medio ordenado | `np.median()` | Centro robusto (no le afectan extremos) |
| Varianza | $\frac{\sum(x_i-\bar{x})^2}{n}$ | `np.var()` | Dispersión (en unidades²) |
| Desv. típica | $\sqrt{\text{Varianza}}$ | `np.std()` | Dispersión (en mismas unidades) |
| Mínimo/Máximo | — | `np.min()` / `np.max()` | Extremos del conjunto |
| Rango | max − min | `np.ptp()` | Amplitud total |
| Percentil P | valor bajo el que cae el P% | `np.percentile(a, P)` | Posición relativa |
| IQR | Q3 − Q1 | `np.percentile(a,75) - np.percentile(a,25)` | Dispersión central |
| Suma acumulada | — | `np.cumsum()` | Totales acumulados |
| `axis=0` | opera por columnas (↓) | `np.mean(a, axis=0)` | Resumir cada columna |
| `axis=1` | opera por filas (→) | `np.mean(a, axis=1)` | Resumir cada fila |
