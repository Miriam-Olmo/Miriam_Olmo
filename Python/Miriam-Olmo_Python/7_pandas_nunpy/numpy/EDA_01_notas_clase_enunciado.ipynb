**Resumen del tema:**

| Concepto | Fórmula | `numpy` | Para qué sirve |
|----------|---------|---------|----------------|
| Media | $\bar{x} = \frac{\sum x_i}{n}$ | `np.mean()` | Valor central típico |
| Mediana | valor del medio ordenado | `np.median()` | Centro robusto (no le afectan extremos) |
| Varianza | $\frac{\sum(x_i-\bar{x})^2}{n}$ | `np.var()` | Dispersión (en unidades²) |
| Desv. típica | $\sqrt{\text{Varianza}}$ | `np.std()` | Dispersión (en mismas unidades) |
| Mínimo/Máximo | — | `np.min()` / `np.max()` | Extremos del conjunto |
| Rango | max − min | `np.ptp()` | Amplitud total |
| Percentil P | valor bajo el que cae el P% | `np.percentile(a, P)` | Posición relativa |
| IQR | Q3 − Q1 | `np.percentile(a,75) - np.percentile(a,25)` | Dispersión central |
| Suma acumulada | — | `np.cumsum()` | Totales acumulados |
| `axis=0` | opera por columnas (↓) | `np.mean(a, axis=0)` | Resumir cada columna |
| `axis=1` | opera por filas (→) | `np.mean(a, axis=1)` | Resumir cada fila |
