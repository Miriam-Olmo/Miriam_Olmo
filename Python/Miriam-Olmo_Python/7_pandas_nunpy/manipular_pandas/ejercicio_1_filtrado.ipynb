# ejercicio 1: filtrado y agrupacion de datos

Trabajas como analista para una plataforma de vídeo. Tienes los datos de visualizaciones de la semana. Debes:

Filtrar el DataFrame para mostrar solo las Series que tengan entre 15 y 50 millones de visualizaciones.
Agrupar los datos por Género, calcular el total de visualizaciones de cada uno y ordenar el resultado de mayor a menor.

```python
datos_streaming = {
    'Título': ['Stranger Things', 'Inception', 'The Crown', 'Avengers', 'Dark', 'Interstellar', 'The Boys'],
    'Tipo': ['Serie', 'Película', 'Serie', 'Película', 'Serie', 'Película', 'Serie'],
    'Género': ['Sci-Fi', 'Sci-Fi', 'Drama', 'Acción', 'Sci-Fi', 'Sci-Fi', 'Acción'],
    'Visualizaciones': [65, 40, 25, 80, 18, 55, 45]
}

df_streaming = pd.DataFrame(datos_streaming)
df_streaming
```

Filtrar el DataFrame para mostrar solo las Series que tengan entre 15 y 50 millones de visualizaciones.

```python
filtro_series = df_streaming[(df_streaming['Tipo'] == 'Serie') & df_streaming['Visualizaciones'].between(15, 50)]
display('series de exito medio', filtro_series)
```

Agrupar los datos por Género, calcular el total de visualizaciones de cada uno y ordenar el resultado de mayor a menor.

# reset_index() genera la visualizacion en formato tabla y un dataframe para poder volver a usar despues






```python
visualizaciones_genero = df_streaming.groupby('Género')['Visualizaciones'].sum().reset_index()
display('Visualización por género', visualizaciones_genero)
```

ordenar el resultado de mayor a menor

```python
order_desc = visualizaciones_genero.sort_values(by='Visualizaciones', ascending=True)
order_desc
```
