# ejercicio 1: filtrado y agrupacion de datos con pandas

dispones de un conjunto de datos de ventas mensuales de distintos empleados. tu objetivo es:
    1- filtrar un dataframe para mostrar solo los vendedores que han facturado mas de 3000€.
    2- agrupar los datos para calcular el total de ventas por region.

```python
# sacar el dataframe e imprimirlo por pantalla
datos = {
    'vendedor': ['Ana', 'Luis', 'Marta', 'Carlos', 'Elena'],
    'region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte'],
    'ventas': [2500, 3200, 4100,1500,3800]
}

df_ventas = pd.DataFrame(datos)

df_ventas
```

1- filtrar ventas superiores a 3000€

```python
ventas_mayor_3000 = df_ventas[df_ventas['ventas'] > 3000]
display('ventas > 3000', ventas_mayor_3000)
```

```python
ventas_entre_1000_2000 = df_ventas[ (df_ventas['ventas'] > 1000) & (df_ventas['ventas'] < 2000)  ]
otro_ventas_1000_2000 = df_ventas[df_ventas['ventas'].between(1000, 2000)]
ventas_entre_1000_2000
otro_ventas_1000_2000
```

2- agrupar los datos para calcular el total de ventas por region

```python
ventas_por_region = df_ventas.groupby('region')['ventas'].sum()
ventas_por_region
```
