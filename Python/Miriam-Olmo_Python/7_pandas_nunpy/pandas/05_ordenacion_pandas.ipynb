# Ordenación y clasificación.
Ordenar es una operación básica pero frecuente. Pandas permite ordenar los dataframe por columnas de cada serie.

```python
import pandas as pd

df = pd.DataFrame({
    "nombre":       ["Ana", "Luis", "Eva", "Carlos", "Marta", "Pedro"],
    "departamento": ["IT", "Ventas", "IT", "RRHH", "Ventas", "IT"],
    "edad":         [23, 31, 28, 45, 22, 38],
    "salario":      [28000, 42000, 35000, 61000, 24000, 55000],
    "nota":         [8.5, 6.0, 9.2, 7.1, 8.8, 6.5]
})


# sort_values - ordenar por valores de columna

print(df.sort_values("salario", ascending=False))


print('-' * 30)

# ordenar por varias columnas, primero ordenamos por departamento asc, luego ordenamos por salario desc 

print( df.sort_values(['departamento', 'salario'], ascending=[True, False]) )
```
